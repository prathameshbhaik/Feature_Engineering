```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('Air_Traffic_Passenger_Statistics.csv',usecols=['Operating Airline',
                                                                  'Operating Airline IATA Code',
                                                                  'Published Airline',
                                                                  'Published Airline IATA Code'])
data.head()
```

```python
data.shape
```

```python
for col in data.columns:
    if data[col].dtype == 'object':
        print(f'{col}:{data[col].unique()}')
        print('========================================================')
```

```python
for col in data.columns:
    print(f'{col} : {data[col].nunique()} labels')
    print('==============')
```

```python
data['Operating Airline'].value_counts().sort_values(ascending=False).head(10)
```

```python
top_10_Operating= [x for x in data['Operating Airline'].value_counts()
                   .sort_values(ascending=False).head(10).index]

top_10_Op_IATA = [x for x in data['Operating Airline IATA Code'].value_counts()
                  .sort_values(ascending=False).head(10).index]

top_10_Published = [x for x in data['Published Airline'].value_counts()
                    .sort_values(ascending=False).head(10).index]

top_10_Pub_IATA = [x for x in data['Published Airline IATA Code'].value_counts()
                   .sort_values(ascending=False).head(10).index]
```

```python
def One_Hot_Enc(df,feature,top_10):
    for label in top_10:
        df[feature+'_'+label] = np.where(df[feature] == label,1,0)
```

```python
One_Hot_Enc(data,'Operating Airline',top_10_Operating)
One_Hot_Enc(data,'Operating Airline IATA Code',top_10_Op_IATA)
One_Hot_Enc(data,'Published Airline',top_10_Published)
One_Hot_Enc(data,'Published Airline IATA Code',top_10_Pub_IATA)

data.drop(['Operating Airline','Operating Airline IATA Code',
               'Published Airline','Published Airline IATA Code'],axis=1,inplace=True)
data.head()
```
